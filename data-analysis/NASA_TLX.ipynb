```python
import pandas as pd
from scipy.stats import ttest_rel, wilcoxon

# Step 1: Define categories and participant scores
categories = ["Mental Demand", "Physical Demand", "Speed Pressure", "Success", "Effort", "Frustration"]

ar_scores = {
    "Mental Demand": [1, 3, 3, 2, 5],
    "Physical Demand": [3, 1, 3, 1, 5],
    "Speed Pressure": [1, 1, 1, 1, 1],
    "Success": [1, 3, 3, 3, 5],
    "Effort": [1, 3, 3, 3, 5],
    "Frustration": [1, 5, 5, 3, 5]
}

vr_scores = {
    "Mental Demand": [2, 1, 3, 3, 1],
    "Physical Demand": [3, 1, 3, 5, 5],
    "Speed Pressure": [1, 3, 1, 1, 1],
    "Success": [3, 3, 3, 1, 3],
    "Effort": [3, 1, 3, 3, 1],
    "Frustration": [3, 3, 5, 3, 1]
}

web_scores = {
    "Mental Demand": [1, 1, 1, 1, 1],
    "Physical Demand": [3, 1, 3, 3, 3],
    "Speed Pressure": [1, 1, 1, 1, 3],
    "Success": [3, 1, 1, 1, 1],
    "Effort": [1, 3, 1, 1, 1],
    "Frustration": [1, 3, 1, 1, 3]
}

# Step 2: Perform t-tests and Wilcoxon tests
results = []

for cat in categories:
    row = {"Category": cat}
    
    for (cond1, cond2, label) in [(ar_scores, vr_scores, "AR vs VR"),
                                  (vr_scores, web_scores, "VR vs Web"),
                                  (ar_scores, web_scores, "AR vs Web")]:
        t_stat, t_p = ttest_rel(cond1[cat], cond2[cat])
        try:
            w_stat, w_p = wilcoxon(cond1[cat], cond2[cat])
        except ValueError:
            w_p = None  # Handle errors if all values are the same

        row[f"{label} t-test p"] = round(t_p, 4)
        row[f"{label} Wilcoxon p"] = round(w_p, 4) if w_p is not None else "NA"

    results.append(row)

# Step 3: Save as CSV
df = pd.DataFrame(results)
df.to_csv("NASA_TLX_Stats_Comparison.csv", index=False)

print("Analysis complete. CSV saved as NASA_TLX_Stats_Comparison.csv")
```
